```r
library(tidyverse)
```

```r
cats1 <- read_delim("cat GPS data/cat_1.csv", ",")
cats2 <- read_delim("cat GPS data/cat_2.csv", ",")
cats3 <- read_delim("cat GPS data/cat_3.csv", ",")
cats4 <- read_delim("cat GPS data/cat_4.csv", ",")
```

```r
library(sf)

cats1 <- cats1 %>% st_as_sf(coords = c("x", "y"), crs = 25833, remove = FALSE)
cats2 <- cats2 %>% st_as_sf(coords = c("x", "y"), crs = 25833, remove = FALSE)
cats3 <- cats3 %>% st_as_sf(coords = c("x", "y"), crs = 25833, remove = FALSE)
cats4 <- cats4 %>% st_as_sf(coords = c("x", "y"), crs = 25833, remove = FALSE)
```

```r
head(cats1)
```

```r
min(cats1$datetime)
max(cats1$datetime)

min(cats2$datetime)
max(cats2$datetime)

min(cats3$datetime)
max(cats3$datetime)
```

```r
# Hier abgekürzt: ohne zuerst die Linie zu berechnen!

ggplot() +
    geom_sf(data = cats1, color = "blue") +
    geom_sf(data = cats2, color = "red") +
    geom_sf(data = cats3, color = "green") +
    geom_sf(data = cats4, color = "yellow")
```

```r
library(leaflet)
leaflet() |>
  addTiles() |>
  addCircleMarkers(data = cats1, color = "red") |>
  addCircleMarkers(data = cats2, color = "blue") |>
  addCircleMarkers(data = cats3, color = "green") |>
  addCircleMarkers(data = cats3, color = "yellow")
```
